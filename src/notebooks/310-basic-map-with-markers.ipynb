## Background map

Let's start by building the background map of our bubble map thanks to the `basemap` library. The `basemap` library allows to load the boundaries of any region of the world without having to load a `shapefile` or a `geojson` file. You can read more about this step in the [map section](https://www.python-graph-gallery/map) of the gallery.

```python
# Load the library
from mpl_toolkits.basemap import Basemap

# Initialize the background map
m=Basemap(llcrnrlon=-160, llcrnrlat=-75,urcrnrlon=160,urcrnrlat=80)

# Draw boundaries, continents and coastlins
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
m.drawcoastlines(linewidth=0.1, color="white")

# show
m
```
